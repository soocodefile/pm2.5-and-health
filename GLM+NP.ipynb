```python

X_train = three_year[['avg_pm']].values  
y_train = three_year['asthma_count'].values 

X_test = year_2014[['avg_pm']]
y_test = asthma_year.iloc[3][['asthma_count']]

k = 3
knn_model = KNeighborsRegressor(n_neighbors=k)
knn_model.fit(X_train, y_train)

y_pred = knn_model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)


print(f"Mean Squared Error: {mse:.2f}")


new_data = np.array(X_test) 
predicted_asthma_cases = knn_model.predict(new_data)
print(f"Predicted asthma cases for 2017 avg_pm: {predicted_asthma_cases[0]:.2f}")
```
