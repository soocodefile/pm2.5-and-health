```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(20, 6))
plt.hist(pm_2014['statefips'], bins=range(pm_2014['statefips'].min(), pm_2014['statefips'].max() + 2), edgecolor='black', align='left')
plt.title('Distribution of StateFIPS in PM2.5 Dataset')
plt.xlabel('StateFIPS')
plt.ylabel('Frequency')
plt.xticks(range(pm_2014['statefips'].min(), pm_2014['statefips'].max() + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
location_counts = asthma_2014['LocationDesc'].value_counts()

# Plot the bar chart
plt.figure(figsize=(10, 6))
location_counts.plot(kind='bar')
plt.title('Distribution of LocationDesc in Asthma 2014 Dataset')
plt.xlabel('LocationDesc (State)')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
asthma_2014 = asthma_2014.copy()

asthma_2014.loc[:, 'DataValue'] = pd.to_numeric(asthma_2014['DataValue'], errors='coerce')

state_asthma_prevalence = asthma_2014.groupby('LocationDesc').agg({
    'DataValue': 'mean'
}).reset_index()

plt.figure(figsize=(12, 6))
plt.bar(state_asthma_prevalence['LocationDesc'], state_asthma_prevalence['DataValue'], color='skyblue')
plt.title('Asthma Prevalence by State (2014)', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Average Asthma Prevalence', fontsize=12)
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
pm_2014['ds_pm_pred'] = pd.to_numeric(pm_2014['ds_pm_pred'], errors='coerce')

pm25_state_avg = pm_2014.groupby('statefips').agg({
    'ds_pm_pred': 'mean'
}).reset_index()

pm25_state_avg['LocationDesc'] = pm25_state_avg['statefips'].map(state_fips)

plt.figure(figsize=(12, 6))
plt.bar(pm25_state_avg['LocationDesc'], pm25_state_avg['ds_pm_pred'], color='lightcoral')
plt.title('PM2.5 Prevalence by State (2014)', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Average PM2.5 (µg/m³)', fontsize=12)
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
def scatterplot(the_year):
    the_year = the_year
    asthma_year = asthma[(asthma['YearStart'] == the_year) & (asthma['YearEnd'] == the_year)].copy()
    pm_2014['ds_pm_pred'] = pd.to_numeric(pm_2014['ds_pm_pred'], errors='coerce')
    asthma_year['DataValue'] = pd.to_numeric(asthma_year['DataValue'], errors='coerce')
    
    pm25_state_avg = pm_2014.groupby('LocationDesc').agg({'ds_pm_pred': 'mean'}).reset_index()
    pm25_state_avg.rename(columns={'ds_pm_pred': 'avg_pm25'}, inplace=True)
    
    asthma_state_avg = asthma_year.groupby('LocationDesc').agg({'DataValue': 'mean'}).reset_index()
    asthma_state_avg.rename(columns={'DataValue': 'avg_asthma_prevalence', 'LocationDesc': 'LocationDesc'}, inplace=True)
    
    merged_data = pm25_state_avg.merge(asthma_state_avg, on='LocationDesc')
    
    plt.figure(figsize=(8, 6))
    plt.scatter(merged_data['avg_pm25'], merged_data['avg_asthma_prevalence'], color='blue', alpha=0.5)
    plt.title('PM2.5 (2014) vs Asthma Prevalence (' + str(the_year) + ')', fontsize=14)
    plt.xlabel('Average PM2.5 (µg/m³)', fontsize=12)
    plt.ylabel('Average Asthma Prevalence', fontsize=12)
    plt.grid(alpha=0.5)
    plt.tight_layout()
    plt.show()
```

Scatterplots PM2.5 (2014) vs. Asthma Prevalence (year between 2014 and 2021)

Trend Observed:
- The points are scattered without a clear linear or non-linear trend. This suggests that there isn’t an apparent direct correlation between PM2.5 levels in 2014 and asthma prevalence in 2014 at the state level. The overall results of scatterplot changed since 2019, but they still not show a strong positive association.

Possible Relationships to Follow Up:
- States with higher PM2.5 concentrations (above 10 µg/m³) do not necessarily show consistently higher asthma prevalence, indicating the relationship may be influenced by other confounding factors.
- Investigate if the asthma prevalence reflects current (2014) exposure to PM2.5 or if there’s a lag effect (e.g., long-term exposure leading to higher prevalence in later years).

Research Question: What is the association between air quality and the development of certain chronic illnesses?

1.	Motivating the Question:
- The scatterplots highlight the complexity of the relationship between air quality and asthma prevalence, motivating further investigation:
- In 2014 (Scatterplot 1), the lack of a clear relationship suggests that asthma prevalence may not be solely driven by current PM2.5 levels, prompting questions about lagged effects or other confounding factors.
- The weak relationship observed in 2021 (Scatterplot 2) encourages studying the long-term impact of PM2.5 exposure on disease prevalence.
3.	Suggesting a Potential Answer:
- The second scatterplot provides some evidence that long-term exposure to higher PM2.5 levels might influence asthma prevalence over time, but the presence of states with low PM2.5 and high asthma prevalence indicates that the relationship is not straightforward.
- Together, the visualizations suggest that PM2.5 levels alone do not fully explain asthma prevalence, highlighting the need for a more comprehensive approach to analyze the role of air quality alongside other factors.
4.	Relevance:
- The visualizations provide a starting point to explore the link between air quality and disease prevalence. They motivate deeper analysis to determine whether PM2.5 has a direct impact, delayed effects, or an indirect association moderated by other variables.

```python
pm_2014['date'] = pd.to_datetime(pm_2014['date'])

# Group by date to calculate daily average PM2.5
pm25_daily_avg = pm_2014.groupby('date')['ds_pm_pred'].mean().reset_index()

# Plot line plot
plt.figure(figsize=(10, 6))
plt.plot(pm25_daily_avg['date'], pm25_daily_avg['ds_pm_pred'], color='blue', linewidth=2)
plt.title('Trends in PM2.5 Levels Over Time (2014)', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Average PM2.5 (µg/m³)', fontsize=12)
plt.grid(alpha=0.5)
plt.show()
```
